## COVID sample data

```python
all_info = pd.read_csv('time_series_covid19_deaths_global.csv')
```

```python
all_info
```

```python
all_info.dropna(subset = ["Lat"],how = "all",inplace=True)
all_info.dropna(subset = ["Long"],inplace=True)
```

```python
all_info['Lat'].isnull().values.any()
```

```python
all_info.head()
```

```python
import folium
world_map = folium.Map()
cases = folium.map.FeatureGroup()

for lat,lng in zip(all_info.Lat,all_info.Long):
    cases.add_child(
        folium.CircleMarker(
        [lat,lng],
        radius=7,
        color='yellow',
        fill=True,
        fill_color='red',
        fill_opacity=0.4
        )
    )
    
map = folium.Map()
map.add_child(cases)
```

```python
cases = folium.map.FeatureGroup()

for lat,lng in zip(all_info.Lat,all_info.Long):
    cases.add_child(
        folium.CircleMarker(
        [lat,lng],
        radius=1,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.4
        )
    )
    
map = folium.Map()
map.add_child(cases)
```

```python
afgh = all_info[all_info['Country/Region'] =='Afghanistan']
df2 = afgh.loc[0]
df2 = df2.drop(labels=['Province/State', 'Country/Region','Lat','Long'])
```

```python
tik = []
for i in range(df2.size):
  tik.append(i)
afgh_df = pd.DataFrame({"days":tik, "people":df2})
afgh_df = afgh_df.astype(int)
```

```python
sns.lineplot(
    data= afgh_df,
    x = afgh_df.days,
    y = afgh_df.people,
)
```

```python
canada_df = all_info[all_info['Country/Region'] =='Canada']
canada_df
```

```python
canada_df = canada_df.sum().drop(labels=['Province/State', 'Country/Region','Lat','Long'])
```

```python
canada = pd.DataFrame({"days":tik, "people":canada_df})
canada = canada.astype(int)
```

```python
sns.lineplot(
    data= canada,
    x = canada.days,
    y = canada.people
)
```
